```
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

rs = 9

campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')

campaign["pdays"] = campaign["pdays"].replace(999, np.NaN)
campaign = campaign.replace("unknown", np.NaN)

campaign_drop_na = campaign.dropna()
```

```
campaign_dummies = pd.get_dummies(campaign_drop_na)

campaign_dummies = campaign_dummies.drop(columns=["y_no"])

campaign_dummies.head()
```

```
features = campaign_dummies.drop(columns=["y_yes"])
targets = campaign_dummies["y_yes"]

X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.3, random_state=rs)

X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=rs)
```

```
#TODO 80 20 split
# Random Forests in `scikit-learn` (with N = 100)
rf = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=rs)

rf.fit(X_train, y_train)
```

```
# This may not the best way to view each estimator as it is small
fn=X_train.columns.values
cn=True
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (20,20), dpi=900)
for index in range(0, 5):
    tree.plot_tree(rf.estimators_[index],
                   feature_names = fn,
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)
fig.savefig('rf_5trees.png')
```

```
y_pred = rf.predict(X_test)
```
